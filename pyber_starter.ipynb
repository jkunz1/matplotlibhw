```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = 'data/city_data.csv'
ride_data_to_load = 'data/ride_data.csv'

# Read the City and Ride Data
ride=pd.read_csv('data/ride_data.csv')
city=pd.read_csv('data/city_data.csv')

# Combine the data into a single dataset
combined_df=pd.merge(ride, city)

# Display the data table for preview
combined_df

citytot=combined_df['city'].value_counts()
combined_df
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban= combined_df[combined_df['type'] == 'Urban'].groupby([combined_df['city']])
rural= combined_df[combined_df['type'] == 'Rural'].groupby([combined_df['city']])
suburban= combined_df[combined_df['type'] == 'Suburban'].groupby([combined_df['city']])

x_urban = urban['ride_id'].count()
y_urban = urban['fare'].mean()
d_urban = urban['driver_count'].mean()

x_rural = rural['ride_id'].count()
y_rural = rural['fare'].mean()
d_rural = rural['driver_count'].mean()

x_suburban = suburban['ride_id'].count()
y_suburban = suburban['fare'].mean()
d_suburban = suburban['driver_count'].mean()

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, label = 'Urban', s=d_urban * 10, color=['lightcoral'], edgecolor='black', alpha = 0.75, marker='o')
plt.scatter(x_rural, y_rural, label = 'Rural', s=d_rural * 10, color=['gold'], edgecolor='black', alpha = 0.75, marker='o')
plt.scatter(x_suburban, y_suburban, label = 'Suburban', s=d_suburban  * 10, color=['lightskyblue'], edgecolor='black', alpha = 0.75, marker='o')
plt.grid()

# Incorporate the other graph properties
plt.title('Pyber Ride Sharing Data 2016')
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')

# Create a legend
legend = plt.legend(fontsize = 8, title= 'City Types') 
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42,35,'Note: Circle size correlates with driver count per city.', fontsize = 10)

# Save Figure
plt.savefig('Images/Pyber_Ride_Sharing_Data_2016.png', bbox_inches='tight')
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
faretype= combined_df.groupby(['type'])['fare'].sum()

# Build Pie Chart
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0,0, 0.1)
plt.pie(faretype, labels = faretype.index, autopct = '%1.1f%%', colors = colors, 
        explode = explode, shadow = True, startangle = 140)
plt.axis('equal')
plt.title('% of Total Fares by City Type')
plt.show()

# Save Figure
plt.savefig('Images/Percent_of_Total_Fares_by_City_Type.png', bbox_inches='tight')
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivertype= combined_df.groupby(['city','type'])['driver_count'].mean()
drivertype2= drivertype.groupby(['type']).sum()
drivertype2

# Build Pie Charts
colors = ['gold', 'lightskyblue', 'lightcoral']
explode = (0,0, 0.1)
plt.pie(drivertype2, labels = drivertype2.index, autopct = '%1.1f%%', colors = colors, 
        explode = explode, shadow = True, startangle = 140)
plt.axis('equal')
plt.title('% of Total Drivers by City Type')

# Save Figure
plt.savefig('Images/Percent_of_Total_Drivers_by_City_Type.png', bbox_inches='tight')
```
